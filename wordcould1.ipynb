```python
# -*- coding:utf-8 -*-  
filename = "text1.txt"
with open(filename) as f:
    mytext = f.read()
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate(mytext)
```

```python
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
